```python
import os
import cv2
import pandas as pd
from deepface import DeepFace
from google.colab.patches import cv2_imshow

# Folder tempat menyimpan foto
folder_path = "foto_wajah"  # ubah sesuai nama folder kamu
output_csv = "hasil_deteksi.csv"

# Pastikan folder ada
if not os.path.exists(folder_path):
    raise FileNotFoundError(f"Folder '{folder_path}' tidak ditemukan!")

# List semua file gambar
image_files = [f for f in os.listdir(folder_path)
               if f.lower().endswith((".jpg", ".jpeg", ".png"))]

if not image_files:
    raise ValueError("Tidak ada file gambar di folder tersebut!")

# Data hasil
hasil_list = []

print("üîç Memulai analisis beberapa foto...\n")

# Loop tiap gambar
for image_name in image_files:
    image_path = os.path.join(folder_path, image_name)
    print(f"üñºÔ∏è Menganalisis: {image_name}")

    try:
        result = DeepFace.analyze(img_path=image_path, actions=["age", "gender"], enforce_detection=False)
        age = result[0]['age']
        gender_dict = result[0]['gender']
        gender_eng = max(gender_dict, key=gender_dict.get)

        # Konversi ke Bahasa Indonesia
        if gender_eng.lower() == "man":
            gender_id = "Laki-laki"
            emoji = "üë¶"
        else:
            gender_id = "Perempuan"
            emoji = "üë©"

        # Simpan ke list hasil
        hasil_list.append({
            "Nama File": image_name,
            "Jenis Kelamin": gender_id,
            "Umur (tahun)": age
        })

        # Tampilkan hasil di terminal
        print(f"   {emoji} Jenis Kelamin : {gender_id}")
        print(f"   üéÇ Perkiraan Umur : {age} tahun\n")

        # (Opsional) tampilkan gambar kecil di Colab
        img = cv2.imread(image_path)
        faces = result[0]['region']
        if faces and faces['w'] > 0:
            x, y, w, h = faces['x'], faces['y'], faces['w'], faces['h']
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
            label = f"{gender_id}, {age} th"
            cv2.putText(img, label, (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2, cv2.LINE_AA)
        # Resize agar kecil
        max_width = 200
        h, w = img.shape[:2]
        if w > max_width:
            scale = max_width / w
            new_dim = (int(w * scale), int(h * scale))
            img = cv2.resize(img, new_dim, interpolation=cv2.INTER_AREA)

        cv2_imshow(img)

    except Exception as e:
        print(f"‚ö†Ô∏è Gagal menganalisis {image_name}: {e}\n")

# Simpan hasil ke CSV
df = pd.DataFrame(hasil_list)
df.to_csv(output_csv, index=False, encoding='utf-8-sig')

print("\n‚úÖ Analisis selesai!")
print(f"üìÅ Hasil tersimpan di file: {output_csv}")
```
